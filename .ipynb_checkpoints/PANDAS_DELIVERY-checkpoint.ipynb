```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
vk = delivery[delivery['batsman']=='CH Gayle']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
def one(b):
    vk = delivery[delivery['batsman']==b]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```
